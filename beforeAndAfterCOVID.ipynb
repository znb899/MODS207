```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import plotly.express as px
import datetime

tqdm.pandas()

pd.set_option('display.max_columns', 500)
```

```python
df = pd.read_pickle('stemmed.pkl')
```

```python
df.head(20)
```

```python
def days(s):
    s = s.split(' days')
    if len(s)==2:
        return int(s[0])
    return 0
```

```python
def toSeconds(a):
    a = str(a)
    a = a.split(':')
    try:
        if len(a)==1:
            return int(a[0])
        if len(a)==2:
            return int(a[0])*60+int(a[1])
        if len(a)==3:
            return int(a[0])*3600+int(a[1])*60+int(a[0])
    except:
        print(a)
```

```python
df.trendingDuration = df.trendingDuration.apply(days)
df['duration'] = df['duration'].apply(toSeconds)
```

```python
df.publishTime = pd.to_datetime(df.publishTime)
df['period'] = '2017-2018'
df.loc[df.publishTime.dt.date>datetime.date(2019, 1, 1), 'period'] = '2020-2022' 
```

```python
title = 'Effect of the duration of the video on the number of trending days'
labels = {'duration':'Duration of the video','period':'Period of time',"trendingDuration":'Number of days trending'}
fig = px.scatter(df[df.trendingDuration>0], x="duration", y="trendingDuration",color='period',labels=labels,title=title, trendline="ols", marginal_y='violin')
fig.show()
```
