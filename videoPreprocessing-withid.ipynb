```python
import pandas as pd
import os
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
# path of trending data
path_trend17_18 = r'C:\Users\Nour\Downloads\FRvideos\FRvideos.csv'
path_trend20_22 = r'C:\Users\Nour\Downloads\archive\FR_youtube_trending_data.csv'
```

```python
trending17_18 = pd.read_csv(path_trend17_18)
trending20_22 = pd.read_csv(path_trend20_22)
```

```python
def preprocess_trenddata(trending17_18, path_trend20_22) :
    trending20_22['trending_date'] = pd.to_datetime(trending20_22['trending_date'].astype(str), format='%Y/%m/%d')
    trending17_18['trending_date'] = pd.to_datetime(trending17_18['trending_date'], format='%y.%d.%m')
    return trending17_18, trending20_22 
```

```python
trending17_18, trending20_22 = preprocess_trenddata(trending17_18, path_trend20_22)
```

```python
# path of all data
path = 'videos_withid'
all_videos = os.listdir(path)
```

```python
def preprocess_data(data):
    data['Publish time'] = pd.to_datetime(data['Publish time'].astype(str), format='%Y/%m/%d')
    # remove the data of 2019
    data17    = data[data['Publish time']>'2017/11/12']
    data17_18 = data17[data17['Publish time']< '2018/06/12']
    data20_22 = data17[data17['Publish time']> '2020/08/09']
    # initialization of trending columns
    trending_times, trending_for = np.zeros(data20_22.shape[0]), np.zeros(data20_22.shape[0])
    data20_22['trending times'] = trending_times
    data20_22['trending for'] = trending_for

    trending_times, trending_for = np.zeros(data17_18.shape[0]), np.zeros(data17_18.shape[0])
    data17_18['trending times'] = trending_times
    data17_18['trending for'] = trending_for
    return data17_18 , data20_22
```

```python
def trending(trendingdata, data):
    for i, title in tqdm(enumerate(data['Title'])):
        trend = trendingdata[trendingdata['title']==title]
        data['trending times'].iloc[i] = trend.count()[0]
        if trend.count()[0] > 0 :
            data['trending for'].iloc[i] = trend['trending_date'].iloc[-1]-trend['trending_date'].iloc[-0] 
    return data 
```

```python
for video in all_videos :
    channel = str("\\") + video
    data = pd.read_csv(path+channel)
    data17_18 , data20_22 = preprocess_data(data)
    data17_18 = trending(trending17_18, data17_18)
    data20_22 = trending(trending20_22, data20_22)
    path_save = 'C:\\Users\\Nour\\all_videos_preprocess_withid'
    data17_18.to_csv(path_save+str("\\")+str('data17_18')+video ,index=False)
    data20_22.to_csv(path_save+str("\\")+str('data20_22')+video ,index=False)
    
```
